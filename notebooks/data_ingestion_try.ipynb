```python
import os

def list_pdfs_in_subfolders(folder_name):
    pdf_dict = {}
    
    # Walk through the directory
    for root, dirs, files in os.walk(folder_name):
        # Check if the current directory is not the root folder
        if root != folder_name:
            pdf_files = [file for file in files if file.lower().endswith('.pdf')]
            if pdf_files:
                # Store the PDF files in the dictionary
                pdf_dict[os.path.basename(root)] = pdf_files
    
    return pdf_dict

# Example usage
folder_name = 'pdf_data'
pdf_lists = list_pdfs_in_subfolders(folder_name)
for subfolder, pdfs in pdf_lists.items():
    print(f"Subfolder: {subfolder}, PDFs: {pdfs}")
```
